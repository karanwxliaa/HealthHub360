```python
df.drop(["Unnamed: 0"], axis =1,inplace = True)
```

```python
n_cols = ['calories', 'cho','Fat', 'pro', 'fluids', 'NA']
```

```python
for col in n_cols :
    x = df[col].quantile(q=[0.20,0.5,0.80],  interpolation='midpoint')

    Q1 =x[0.20]
 
    Q3 = x[0.80]
    IQR = Q3 - Q1
    repl = df[col].mean()
    df[col].fillna(repl,inplace=True)
    df.loc[df[col]>= (Q3+1.5*IQR), col] = round(repl,1)
    df.loc[df[col]<= (Q1-1.5*IQR), col] = round(repl,1)
 
    
 
```
